```python
from fetcher import get_player_form
import pandas as pd
```

```python
LAST_GAMEWEEK = 4
```

```python
players = get_player_form(LAST_GAMEWEEK)
```

```python
players_df = pd.DataFrame(players)

players_df['xG'] = players_df['xG'].astype(float)
players_df['xA'] = players_df['xA'].astype(float)
players_df['key_passes'] = players_df['key_passes'].astype(int)
players_df['shots'] = players_df['shots'].astype(int)
players_df['goals'] = players_df['goals'].astype(int)
players_df['assists'] = players_df['assists'].astype(int)
players_df['npxG'] = players_df['npxG'].astype(float)
players_df['xG+A'] = players_df['xG'] + players_df['xA']
```

```python
# Get Top 5 based on Attacking Point per Club
players_df['attacking_point'] = (players_df['goals'] * 5) + (players_df['assists'] * 3)
players_df_sort = players_df.sort_values(by='attacking_point', ascending=False).groupby('team_title').head(5)
players_df_sort = players_df_sort[(players_df_sort.attacking_point >= 10)]
players_df_sort = players_df_sort[(players_df_sort.npxG >= 1)]
# players_df_sort = players_df_sort.sort_values(by=['team_title','attacking_point'], ascending=[True,False])
players_df_sort[['player_name','team_title', 'attacking_point', 'goals', 'assists', 'shots', 'npxG', 'key_passes', 'games', 'time']]
```

```python
# Get Potential based on Attacking Point per Club
players_df_potential = players_df.drop(players_df_sort.index)
```

```python
players_df_potential_sort = players_df_potential.sort_values(by='xG', ascending=False).groupby('team_title').head(2)
players_df_potential_sort = players_df_potential_sort[(players_df_potential_sort.attacking_point > 0)]
players_df_potential_sort = players_df_potential_sort[(players_df_potential_sort.xG > 1.2)]
players_df_potential_sort = players_df_potential_sort[(players_df_potential_sort.shots >= 10)]
players_df_potential_sort = players_df_potential_sort.sort_values(by=['team_title','xG'], ascending=[True,False])
players_df_potential_sort[['player_name','team_title', 'xG', 'goals', 'assists', 'shots', 'games', 'time']]
```

```python
players_df_def = players_df[(players_df.position.str.contains("D"))]
players_df_def = players_df_def[(players_df_def.attacking_point > 0)]
players_df_def = players_df_def[(players_df_def.shots > 2)]
players_df_def = players_df_def.sort_values(by=['shots', 'xG'], ascending=[False, False])
players_df_def[['player_name','team_title', 'position', 'xG', 'goals', 'assists', 'shots', 'games', 'time']]
```

```python
players_df_def = players_df[(players_df.position.str.contains("D"))]
players_df_def = players_df_def[(players_df_def.xG >= 1)]
players_df_def = players_df_def[(players_df_def.shots > 3)]
players_df_def = players_df_def.sort_values(by=['xG'], ascending=[False])
players_df_def[['player_name','team_title', 'position', 'xG', 'goals', 'assists', 'shots', 'games']]
```

```python
# Sort player per team
players_df_club = players_df[(players_df.team_title == "Arsenal")]
players_df_club = players_df_club[(players_df_club.xG >= 0.6)]
players_df_club = players_df_club.sort_values(by=['shots'], ascending=[False])
players_df_club[['player_name','team_title', 'position', 'attacking_point', 'xG', 'goals', 'assists', 'shots', 'key_passes', 'games', 'time']]
```
